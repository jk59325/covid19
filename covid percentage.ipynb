```python
# convert all cells with percentage sign to floats
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

def percConv(x):
    #print (type(x))
    if (type(x) is str):
        if '%' in x:
            #print('hit')
            x= x.replace('%','')
            x= float(x)    
    return x

#percConv("9.35%")
new = new.applymap(percConv)


             
new
```

```python
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

def getLast5(rownum):
    row0 = new.iloc[rownum]
    last5 = row0[-5:]
    #last5 = last5.apply( lambda x : x.replace('%',''))
    #last5 = pandas.to_numeric(last5,errors='coerce',downcast='float')
    #last5.astype(float)
    return last5

ROWS,COLS = new.shape
length =5
x = np.arange(length, dtype=float).reshape((length, 1))
slopeList = list()
for r in range(ROWS):
    #print(r)
    y = getLast5(r)
    #x = x.reshape(length, 1)
    #y = y.reshape(length, 1)
    regr = linear_model.LinearRegression()
    regr.fit(x,y)
    #print(regr.coef_)
    slopeList.append(regr.coef_[0])
    #print(last5)

#print(slopeList)
df_slope = pandas.DataFrame({'slope':slopeList})
```

```python
df = pandas.concat([new,df_slope],axis=1)
df = df.sort_values(by=['slope'])
df
```

```python
tp = df[-5:]
#print(topPerf)
tp = tp.reset_index()
tp = tp.drop("index",1)
tp = tp.drop("slope",1)
tp.rename({'STATE OF REOPENING':'STATE'}, axis=1,inplace=True)
tp = tp.set_index("STATE")
tp = tp.T


    
print(tp)
```
